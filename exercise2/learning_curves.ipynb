# 2. Plot Learning Curves

```python
import json
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Plot learning curves of different learning rates.
plt.figure(1, (8, 6))
for i in range(1, 5):
    with open('./results/results_run_' + str(i) + '.json') as f:
        result = json.load(f) # this is dictionary
        learning_curve = result['learning_curve']
        plt.plot(learning_curve, label=str(result['lr']))
    plt.legend(loc='best')
    plt.title('Learning curves with different learning rates')
    plt.xlabel('epoch')
    plt.ylabel('validation accuracy')
    plt.grid(True)    
```

```python
# Plot learning curves of different filter sizes. I could only run with filter size 1 and 3. If I set it to 5 and higher,
# it always results in memory overflow with the provided pool computers.
plt.figure(2, (8, 6))
#for i in range(5, 9):
for i in range(5, 7):
    with open('./results/results_run_' + str(i) + '.json') as f:
        result = json.load(f)
        learning_curve = result['learning_curve']
        plt.plot(learning_curve, label='filter size: %i' % result['filter_size'])
    plt.legend(loc='best')
    plt.title('Learning curves with different filter sizes')
    plt.xlabel('epoch')
    plt.ylabel('validation accuracy')
    plt.grid(True)
```

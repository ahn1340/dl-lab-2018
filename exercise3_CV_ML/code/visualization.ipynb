## Visualizing the IOU curves

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Plot the IoU curves of different configurations.

for i in range(1, 5):
    iou_path = "./checkpoints/config" + str(i) + "/testIoU.txt"
    file = open(iou_path, "r")
    testIoU = []
    epochs = []
    for line in file:
        #print(line)
        epoch, IoU_score = line.split(" ")
        epochs.append(int(epoch)/500)
        testIoU.append(float(IoU_score))
    plt.plot(epochs, testIoU, label="config" + str(i))

    # Print best IoU values for each configuration.
    print("Best IoU for configuration %i: %.4f" % (i, np.max(testIoU)))

# Plot.
plt.title("TestIOU")
plt.xlabel("Epoch")
plt.ylabel("IoU")
plt.legend(["config1", "config2", "config3", "config4"])
plt.show()
```
